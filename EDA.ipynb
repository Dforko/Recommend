```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
pd.set_option("max_colwidth", 270)
db= pd.read_parquet(r'dbOK_small.parquet')
```

```python
db.info()
```

```python
db=db.drop_duplicates(subset=['overview', 'title'])
```

```python
def is_cjk_arabic_hebrew_russian(s):
    return bool(re.search(r'[\u4E00-\u9FFF\u3040-\u30FF\u0600-\u06FF\u0590-\u05FF\u0400-\u04FF]', s))

# Drop rows where the 'title' column contains Chinese, Japanese, Arabic, Hebrew or Russian characters
db= db[~db['overview'].apply(is_cjk_arabic_hebrew_russian)]



```

```python
def is_cjk_arabic_hebrew_russian_polish(s):
    return bool(re.search(r'[\u4E00-\u9FFF\u3040-\u30FF\u0600-\u06FF\u0590-\u05FF\u0400-\u04FF\u0100-\u017F]', s))
# Drop rows where the 'title' column contains Chinese, Japanese, Arabic, Hebrew, Russian or Polish characters
db = db[~db['overview'].apply(is_cjk_arabic_hebrew_russian_polish)]
```

```python
db = db[~db['title'].apply(is_cjk_arabic_hebrew_russian_polish)]
```

```python
result = db[db['overview'].apply(is_cjk_arabic_hebrew_russian)].sort_values(by='title')
print(result)
```
